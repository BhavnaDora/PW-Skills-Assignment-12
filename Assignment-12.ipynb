##### The errors defined in ArithmeticError class are ZeroDivisonError, FloatingPointError, OverFlowError

```python
#ZeroDivisionError-
import logging
try:
    a=10
    a/0
except ZeroDivisionError as e:
    logging.basicConfig(filename= "test.log", level= logging.ERROR)
    logging.error(e)
    
```

```python
#OverflowError-
try:
    f=3.0
    for i in range(100):
        print(i, f)
        f = f ** 2
except OverflowError as e:
    print('Overflowed after ', f, e)
```

#### (4)-
##### LookupError class is used to catch the exceptions that occur when key or index used for mapping a dictionary or sequence does not exist or is invalid.
##### Key error- when the key we are trying to access is not found in the dictionary, a key error exception is raised
##### Index error- when we are trying to access an index of a list/sequence which does not exists or is out of range, then index error is raised.

```python
# index error-
import logging
try:
    a= [10,46,"hello",(2,3,1),"pwskills"]
    a[7]
except LookupError as e:
    logging.basicConfig(filename= "test1.log", level= logging.ERROR)
    logging.error(e)
```

```python
# key error- 
try:
    d= {"name":"bhavna", "age":24, "course":"DSM"}
    user_input= input("enter any info you wish to find")
    print("in the dictionary {}".format(d[user_input]))
except LookupError as e:
    print("key error is raised",e)
```
